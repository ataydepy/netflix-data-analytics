```python
import pandas as pd

# Caminho direto e definitivo que encontramos
caminho = r"C:\Users\Ataide\.cache\kagglehub\datasets\sayeeduddin\netflix-2025user-behavior-dataset-210k-records\versions\1\movies.csv"

# Carrega os dados para o DataFrame
df = pd.read_csv(caminho)

# Exibe as 5 primeiras linhas para começarmos a analisar
df.head()
```

```python
# Mostra o total de (linhas, colunas) do seu banco de dados
df.shape
```

```python
# Mostra quantos valores nulos (vazios) existem em cada coluna
df.isnull().sum()
```

```python
# 1. Se for filme (ou não tiver temporadas/episódios), preenche com 0
df['number_of_seasons'] = df['number_of_seasons'].fillna(0)
df['number_of_episodes'] = df['number_of_episodes'].fillna(0)

# 2. Se não tiver gênero secundário, preenche com 'Nenhum'
df['genre_secondary'] = df['genre_secondary'].fillna('Nenhum')

# 3. Vamos verificar se as limpezas funcionaram!
df[['number_of_seasons', 'number_of_episodes', 'genre_secondary']].isnull().sum()
```

```python
# 1. Calcula a mediana das notas do IMDb (geralmente fica perto de 6.5)
mediana_imdb = df['imdb_rating'].median()

# 2. Preenche os nulos do IMDb com essa mediana
df['imdb_rating'] = df['imdb_rating'].fillna(mediana_imdb)

# 3. Verifica se ainda sobrou algum nulo no IMDb
print("Nulos no IMDb:", df['imdb_rating'].isnull().sum())
```

```python
# Salva o banco de dados limpo em um novo arquivo chamado 'movies_limpo.csv'
df.to_csv('movies_limpo.csv', index=False)
print("Dataset limpo e salvo com sucesso!")
```
